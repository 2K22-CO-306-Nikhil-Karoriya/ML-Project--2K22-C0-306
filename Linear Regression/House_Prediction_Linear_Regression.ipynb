**House Prediction Model** : Linear Regression

We will use california dataset instead of Boston because target values were missing in that dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
```

```python
Hdata= fetch_california_housing()
Hdata
```

The attributes we will need are the following:

1) feature_names

2)data

3) target

```python
x= Hdata.data
y= Hdata.target
X= Hdata.feature_names
dsc= Hdata.DESCR
```

```python
df= pd.DataFrame(x)
df.columns= X
print(df.head(10))
```

```python
print(dsc)
```

```python
# Splitting our data in training set and testing set using scikit learn
from sklearn import model_selection
print(model_selection.train_test_split(x,y))
x_train, x_test, y_train, y_test=model_selection.train_test_split(x,y)
```

# Training Linear Regression Model

```python
# Importing Linear Regression model from SKlearn and training model on our training dataset
from sklearn.linear_model import LinearRegression
LR= LinearRegression()
LR.fit(x_train,y_train)
y_prediction= LR.predict(x_test)
print(y_prediction)
```

# Plotting Graph

```python
plt.plot([0,1,6],[0,1,6], 'r-')
plt.axis([0,6,0,6])
plt.title("Prediction value vs Actual value")
plt.scatter(y_prediction, y_test, s=.1)
plt.xlabel('y_prediction')
plt.ylabel('y_test')
plt.show()
```

```python
LR.score(x_train,y_train)
```
